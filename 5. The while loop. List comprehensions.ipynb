# Homework: The while loop. List comprehensions


## Task 1

```python
counter = -3

while counter <= 5:
    print(counter)
    counter += 1
```

## Task 2

```python
numbers = [11, 4, 9, 16, 7, 5]

squares = [number ** 2 for number in numbers]
print(f"Squares of all numbers: {squares}")

even_numbers = [number for number in numbers if number % 2 == 0]
print(f"Only even numbers: {even_numbers}")

multiplied_by_10 = [number * 10 for number in numbers]
print(f"All numbers multiplied by 10: {multiplied_by_10}")
```
