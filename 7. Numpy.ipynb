## Task 2

```python
arr = np.array([12, 24, 36, 48, 60])
print(arr)

mult = arr * 2
print("multiplied by 2:", mult)

add = arr + 10
print("added 10:", add)

square = arr ** 2
print("squared elements:", square)

sqrt = np.sqrt(arr)
print("square root:", sqrt)
```

## Task 3

```python
data = np.array([18, 27, 33, 41, 52, 68, 79])
print(data)

mean = np.mean(data)
print("mean:", mean)

min_v = np.min(data)
print("minimum:", min_v)

max_v = np.max(data)
print("maximum:", max_v)

sum_v = np.sum(data)
print("sum:", sum_v)
```
