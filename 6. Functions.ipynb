# Task 1

```python
def show_menu(dishes):
    print("=== MENU ===")
    for index, dish in enumerate(dishes, start = 1):
        print(f"{index}. {dish}")
    print("=============")
    
menu_items = ["Pasta", "Pizza", "Tiramisu"]
show_menu(menu_items)
```
