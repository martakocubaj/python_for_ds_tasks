# Task 1


```python
cooking_time = 5
if cooking_time < 4:
    print("Egg is runny")
elif cooking_time < 7:
    print("Soft-boiled egg")
elif cooking_time < 10:
    print("Hard-boiled egg")
else: 
    print("Overcooked egg") 
```

## Task 2

```python
grade = 72

if grade >= 90:
    print("Excellent")
elif 75 <= grade:
    print("Good")
elif 60 <= grade:
    print("Satisfactory")
else:
    print("Unsatisfactory")
```

## Task 3

```python
refrigerator = ["milk", "eggs", "butter", "cheese"]
```

```python
def find_in_refrigerator(find):
    if find in refrigerator:
        print(f"{find} in the refrigerator")
    else:
        print(f"no {find} in the refrigerator")
```

```python
find_in_refrigerator("eggs")
find_in_refrigerator("bread")
```
