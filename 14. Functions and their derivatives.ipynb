1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
f_x = 10 * x - 5
f_der = np.full_like(x, 10)

plt.plot(x, f_x, label='f(x) = 10x - 5')
plt.plot(x, f_der, label="f'(x) = 10")
plt.legend()
plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp

x = sp.symbols('x')
f_sin = sp.sin(x)
der_sin = sp.diff(f_sin, x) 
print(der_sin)
```

```python
f_sin_func = sp.lambdify(x, f_sin, 'numpy')
der_sin_func = sp.lambdify(x, der_sin, 'numpy')
```

```python
x = np.linspace(-5, 5, 200)
plt.figure(figsize=(10, 5))
plt.plot(x, f_sin_func(x), label='sin(x)')
plt.plot(x, der_sin_func(x), label='cos(x)', linestyle='--')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = sp.symbols('x')
f_ln = sp.ln(x)
der_ln = sp.diff(f_ln, x)
print(der_ln)
```

```python
f_ln = sp.lambdify(x, f_ln, 'numpy')
der_ln = sp.lambdify(x, der_ln, 'numpy')
```

```python
x = np.linspace(-2, 2, 100)
plt.figure(figsize=(10, 5))
plt.plot(x, f_ln(x), label='ln(x)')
plt.plot(x, der_ln(x), label='1/x', linestyle='--')
plt.legend()
plt.grid(True)
plt.show()
```
